#長照 2.0 暖心神隊友小林：基於 RAG 檢索增強技術之照護者支持系統
Smart Support System for Caregivers: A RAG-Based Approach to Long-Term Care 2.0 Resources

##一、 專案動機 (Project Motivation)
1. 照顧者的資訊焦慮與「隱形負擔」
在長照 2.0 的政策架構下，雖然政府提供了豐富的資源，但資訊散落在數百頁的官方 PDF 手冊與繁瑣的法規條文中。對於處於高壓環境、飽受「休息焦慮（Rest Anxiety）」與身心疲憊困擾的家庭照顧者而言，「尋找正確資訊」本身就是一種沉重的隱形負擔。
2. 官方文獻的「可及性」困境
傳統的搜尋方式（如關鍵字檢索）往往無法精準回應使用者的口語化問題。雖然通用型 AI（如 ChatGPT）能回答問題，但也可能因為使用者的prompt形成不同的回答，導致錯誤的政策引導或補助計算。

3. 從「冷冰冰的資訊」到「溫暖的支持」
目前的自動化諮詢系統多半語氣生硬，缺乏對照顧者心理狀態的理解。本專案不僅旨在提供技術性的解答，更試圖透過 AI 人設的刻畫，在提供正確法規資訊的同時，給予照顧者最基礎的情感共鳴與支持。

##二、 專案核心價值與目標 (Core Values & Objectives)
1. 實踐「資訊平權」：
利用 RAG (Retrieval-Augmented Generation) 技術，將長照 2.0 核心官方手冊轉化為智慧知識庫。透過語意搜尋，讓複雜的法規和文件變成人人都能聽懂、隨問隨答的口語化建議。

2. 擴充「地區性資料」:
將抽象的政策轉化為家門口就能找到的具體幫助。以桃園市作為擴充示範。

3. 確保「絕對精準」：
透過現代化的 LCEL (LangChain Expression Language) 架構，強制 AI 的回答必須「本於事實（Fact-grounded）」。系統會精準檢索文本切片（Chunks），並附上參考來源，確保每筆補助資訊、申請流程皆有據可循。

4. 建構「情感防線」：
透過提示詞工程 (Prompt Engineering) 賦予 AI 「暖心支持專員」的角色。每一則回覆均以同理照顧者為出發點，將 AI 的功能從「資訊檢索」提升至「陪伴與導航」，真正實現科技服務於人的終極目標。

```python
# load data and split data
import os
from langchain_community.document_loaders import PyMuPDFLoader, CSVLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

# 1. 定義資料資料夾路徑
data_folder = "./data"
all_documents = []

# 2. 自動偵測格式並載入檔案
print(" 開始掃描 data 資料夾...")

for filename in os.listdir(data_folder):
    file_path = os.path.join(data_folder, filename)

    # 處理 PDF 檔案
    if filename.endswith(".pdf"):
        loader = PyMuPDFLoader(file_path)
        all_documents.extend(loader.load())
        print(f" 已成功載入 PDF: {filename}")

    # 處理 CSV 檔案
    elif filename.endswith(".csv"):
        # 建議指定 encoding='utf-8' 以確保中文不亂碼
        # 如果你的 CSV 是從 Excel 匯出的，可能需要改為 encoding='big5'
        loader = CSVLoader(file_path, encoding='utf-8')
        all_documents.extend(loader.load())
        print(f" 已成功載入 CSV: {filename}")

# 3. 文本切割 (Text Splitting)
# 對於 CSV 來說，通常每一列就是一個 Chunk，但為了與 PDF 統一進入向量資料庫，我們使用相同的切割器
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,        # 每個片段的大小
    chunk_overlap=100,     # 片段間的重疊長度
    separators=["\n\n", "\n", "。", "！", "？", " ", ""]
)

all_splits = text_splitter.split_documents(all_documents)

print("-" * 30)
print(f" 讀取完成！")
print(f" 原始檔案總數: {len(os.listdir(data_folder))} 個")
print(f" 切割後的片段總數: {len(all_splits)} 個")
```

```python
# vector embedding
import time
import os
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import Chroma

# 1. 設定模型
embeddings = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")
db_path = "./chroma_db"

# 2. 載入現有的資料庫
if os.path.exists(db_path):
    print(f" 偵測到現有資料庫，正在讀取進度...")
    vectorstore = Chroma(persist_directory=db_path, embedding_function=embeddings)
    # 檢查目前資料庫裡已經有多少片段
    already_done = len(vectorstore.get()['ids'])
    print(f" 目前進度：{already_done}/{len(all_splits)}")
else:
    print(" 資料庫不存在，將從頭開始建立...")
    already_done = 0
    # 初始化
    vectorstore = Chroma.from_documents(
        documents=all_splits[:1],
        embedding=embeddings,
        persist_directory=db_path
    )
    already_done = 1

# 3. 只處理剩下的片段
BATCH_SIZE = 10
for i in range(already_done, len(all_splits), BATCH_SIZE):
    batch = all_splits[i : i + BATCH_SIZE]

    while True:
        try:
            vectorstore.add_documents(batch)
            current_count = min(i + BATCH_SIZE, len(all_splits))
            print(f" 補完進度: {current_count}/{len(all_splits)} 已完成")
            time.sleep(10) # 稍微拉長休息時間，穩定連線
            break
        except Exception as e:
            print(f" 遇到錯誤 ({e})，休息 30 秒後重試...")
            time.sleep(30)

print("\n 恭喜！所有 603 個片段已全部向量化完畢！")
```

```python
# build RAG chain
# 根據metric調整後的model
import os
from langchain_google_genai import ChatGoogleGenerativeAI, GoogleGenerativeAIEmbeddings
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
from langchain_community.vectorstores import Chroma

# 1. 載入資料庫與 Embedding
db_path = "./chroma_db"
embeddings = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")
vectorstore = Chroma(persist_directory=db_path, embedding_function=embeddings)
retriever = vectorstore.as_retriever(search_kwargs={"k": 4})

#  【新增】格式化函數：將內容與來源檔名結合
def format_docs_with_sources(docs):
    formatted_parts = []
    for i, doc in enumerate(docs):
        # 從 metadata 中取得檔名，並過濾掉路徑只留名稱
        source_name = os.path.basename(doc.metadata.get('source', '未知來源'))
        formatted_parts.append(f"--- 參考片段 {i+1} (來自: {source_name}) ---\n{doc.page_content}")
    return "\n\n".join(formatted_parts)

# 2. 定義提示詞 (強化版：要求標註來源與口語化)
system_prompt = (
    "你是一位溫暖專業的「長照 2.0 照顧者支持專員」。\n"
    "開頭請務必先說：「辛苦了，這段照顧的路不容易，我們在這裡陪你。」\n\n"
    "【回答準則】\n"
    "1. 請嚴格根據下方提供的「參考資料」回答。若資料不足，請誠實告知並建議詢問 1966 長照專線。\n"
    "2. 請將生硬的法律或機構資訊轉化為「民眾聽得懂的白話文」。\n"
    "3. **重要**：在回答的最末尾，請條列出你參考的檔案名稱（例如：參考來源：桃園市日間照顧中心名冊）。\n\n"
    "【參考資料】\n"
    "{context}"
)

prompt = ChatPromptTemplate.from_messages([
    ("system", system_prompt),
    ("human", "{question}"),
])

# 3. 初始化大腦 (Gemini 2.5 Flash)
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0.3)

# 4. 建立現代化 RAG 鏈 (整合格式化函數)
rag_chain = (
    {"context": retriever | format_docs_with_sources, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)



# 5. 啟動對話介面
print(" 「長照 2.0 照顧者神隊友」來源標註版已上線！")
print("-" * 50)

while True:
    query = input(" 您想詢問什麼？")
    if query.lower() in ['exit', '離開']:
        print(" 小林：記得休息，我們下次見！")
        break

    if not query.strip(): continue

    print(" 搜尋手冊中...")
    try:
        # 直接執行鏈
        response = rag_chain.invoke(query)
        print(f"\n 小林回答：\n{response}")
        print("-" * 50)
    except Exception as e:
        print(f" 出現問題：{e}")
```

#Metric
##RAG 評估
1. 忠實度 (Faithfulness)：AI 的回答是否「完全」來自文本切片？

2. 答案相關性 (Answer Relevance)：AI 是否正面回答了照顧者的問題？例如家屬問補助問題，AI要給出具體計畫

3. 檢索精確度 (Context Precision)：在檢索出的 4 個片段中，真正包含答案的比例有多少？評估你的向量資料庫與 Embedding 是否匹配良好。

##系統穩定性與效能
1. 平均延遲 (Latency)：從使用者按下 Enter 到看到第一個字需要多久？

2. 來源引用透明度 (Source Attribution Rate)：系統在回答中正確標註參考檔案的比例。這是打破黑盒子的關鍵。

##民眾實用
1. 用戶體驗與情感支持 (UX & Empathy Score)
- 同理心感受度 (Empathy Rating)：透過測試者評分，評估 AI 的語調與說話方式是否能有效降低使用者在諮詢時的焦慮感。

2. 語言易讀性 (Readability)
- 白話轉譯率：將生硬的法律條文（如獎助計畫經費基準）轉換為民眾聽得懂的口語的成功程度。

1.  **居家服務的定義**：
    居家服務屬於「居家式」的長照服務，簡單來說，就是由專業人員到您的家中提供所需的照顧服務。

2.  **申請流程**：
    *   您需要向「照管中心」（長期照顧管理中心）或您所在縣市的主管機關提出申請。
    *   他們會對您的狀況進行評估，並根據評估結果來提供服務。

3.  **若涉及醫療照護**：
    如果居家服務的內容包含醫事照護（例如護理服務），那麼除了照管中心的評估外，還需要有醫師開立的意見書。

4.  **費用與補助**：
    *   照管中心或縣市主管機關會根據您的失能程度，核定您的「長照需要等級」和「長照服務給付額度」（也就是政府會補助的服務費用上限）。
    *   您需要自行負擔核定金額中的一部分比例或金額。
    *   政府會依據失能者的失能程度和家庭經濟狀況提供補助。但請注意，如果其他法令也有提供相同性質的服務補助，您只能選擇其中一種來申請。
    *   長照特約單位（提供服務的機構）會依照規定向您收取您應自行負擔的費用，這部分是不能減免的。
